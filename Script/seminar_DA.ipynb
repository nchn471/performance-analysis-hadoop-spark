# Visualize

```python
import matplotlib.pyplot as plt

titles = {
    "input_split_mb": "Input Split",
    "buffer_kb": "Shuffle",
    "max_size_in_flight_mb": "Shuffle",
    "parallelism" : "Shuffle"
    }
legends = {
    "input_split_mb": "Split",
    "buffer_kb": "Buffer",
    "max_size_in_flight_mb": "MaxSizeInfo",
    "parallelism" : "PL"  
}
y_labels = {"duration_s": "Execution Time (sec)",
            "cpu_used_percent" : "CPU Used (percent)",
            "ram_used_mb" : "Ram Used (MB)"}

def visualize(df1,df2=None, cat=["input_split_mb"], y_axis="duration_s"):
    if isinstance(cat, str):
        cat = [cat]

    x = [1, 5, 10, 15, 20, 25, 30]  

    plt.figure(figsize=(10, 6))
    for df in df1, df2:
        if df is None:
            continue
        if len(cat) == 1:
            for cat_value in df[cat[0]].dropna().unique():
                y = list(df[y_axis][df[cat[0]] == cat_value])
                plt.plot(x, y, marker='*', label=f'{legends[cat[0]]}-{cat_value}, Exc {df.iloc[0]["num_executors"]}, EC {df.iloc[0]["executor_cores"]}, EM {df.iloc[0]["executor_memory_mb"]}, {df.iloc[0]["num_nodes"]} node(s)')
        else:
            for group_values, group_data in df.groupby(cat):
                y = list(group_data[y_axis])
                label = f"{legends[cat[0]]}-{group_values[0]},{legends[cat[1]]}-{group_values[1]}, Exc {df.iloc[0]["num_executors"]}, EC {df.iloc[0]["executor_cores"]}, EM {df.iloc[0]["executor_memory_mb"]}, {df.iloc[0]["num_nodes"]} node(s)"
                plt.plot(x, y, marker='*', label=label)
    
    if df2 is not None:
        plt.title(f"{titles[cat[0]]} ({df1.iloc[0]['type']})", fontsize=14)
    else:
        plt.title(f"{titles[cat[0]]} ({df1.iloc[0]['type']}) - {df1.iloc[0]["num_nodes"]} Datanode(s)", fontsize=14)

    plt.xlabel("Data Size (GB)", fontsize=12)
    plt.ylabel(f"{y_labels[y_axis]}", fontsize=12)
    plt.xticks(x)  
    plt.legend(fontsize=10)
    plt.show()
```

```python
def visualize_speed_up(df1, df2, split_size):
    
    df1_speed_up = df1.groupby(['input_datasize_bytes', 'input_split_mb'])['duration_s'].mean().reset_index()
    df2_speed_up = df2.groupby(['input_datasize_bytes', 'input_split_mb'])['duration_s'].mean().reset_index()

    df_speed_up = pd.merge(df1_speed_up, df2_speed_up, on=['input_datasize_gb', 'input_split_mb'], suffixes=('_1', '_2'))

    df_speed_up['speed_up'] = df_speed_up['duration_s_1'] / df_speed_up['duration_s_2']

    x = [1, 5, 10, 15, 20, 25, 30]

    y = list(df_speed_up['speed_up'][df_speed_up['input_split_mb']==split_size])

    ax = plt.gca()
    ax.plot(x, y, marker='*', label=f"Split-Size-{split_size}")

    ax.set_xlim(1) 

    ax.set_title(f"{df1.iloc[0]['type']} Speedup of {df1.iloc[0]['num_nodes']} datanode(s) over {df2.iloc[0]['num_nodes']} datanode(s)", fontsize=14)
    ax.set_xlabel("Data Size (GB)", fontsize=12)
    ax.set_ylabel("Speed up", fontsize=12)
    ax.set_xticks(x) 
    ax.legend(fontsize=10)    
```

```python
# Vẽ biểu đồ
plt.figure(figsize=(14, 5))  # Tạo một canvas có kích thước lớn hơn

# Biểu đồ 1: 1 datanode
plt.subplot(1, 2, 1)
visualize_speed_up(wordcount1, wordcount2, 128)

# Biểu đồ 2: 2 datanodes
plt.subplot(1, 2, 2)
visualize_speed_up(terasort1, terasort2, 256)

# Hiển thị biểu đồ
plt.tight_layout()  # Tự động điều chỉnh khoảng cách giữa các biểu đồ
plt.show()
```
